# 3- How to rename Columns?

```python
df = pd.DataFrame({"A":[1,2,3],"B":[4,5,6] })
df
```

```python
df.rename(columns={"A":"col_a","B":"col_b"}, inplace = True)
df
```

```python
df.columns ={'column_a', 'column_b'}
df
```

```python
# to replace column name
df.columns = df.columns.str.replace('_',' ') 
df
```

```python
df.columns = df.columns.str.replace(' ','_') 
df
```
