```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5,0.1,0.8,0.8] ) # from this we can control the left, bottom, width, height of the canvas(from 0 to 1)
axis.plot(x,y, 'r')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 3) # submap is of one row and two columns
for ax in axes:
    ax.plot(x,y, 'r')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1) # submap is of one row and two columns
for ax in axes:
    ax.plot(x,y, 'r')
```

```python
# we can also draw a picture or graph inside a graph
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8] ) # big axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3] ) # small axes

axes1.plot(x,y,'r')
axes2.plot(x,y,'g')
```

```python
fig = plt.figure() # new graphics objects
fig.add_subplot()
plt.plot(x,y,'r')
```

```python
import numpy as np
x = np.linspace(0, 10, 25)
y = x*x+2
print(x)
print(y)
```

```python
fig = plt.figure(figsize=(16,9), dpi = 300) # new graphics objects
fig.add_subplot()
plt.plot(x,y,'r')
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'r')
axes.set_title('title')
axes.set_xlabel('x label')
axes.set_ylabel('y label')
```

```python
ax.legend('label1', 'label2')


fig, axes = plt.subplots(dpi= 200)

axes.set_xlabel('x label')
axes.set_ylabel('y label')

axes.set_title('title')
axes.plot(x, x**2)
axes.plot(x, x**3)

axes.legend(["y = x**2", "y= x**3"], loc=2)
```

```python
# in matplotlib you can set light colours and transperancy and much more
```

```python
fig, axes = plt.subplots(dpi= 150)
axes.plot(x,x+1, color = 'red')
axes.plot(x,x+2, color = 'blue')
axes.plot(x,x+3, color= 'green')
```

```python
fig, axes = plt.subplots(dpi= 150)
axes.plot(x,x**2, color = 'red', alpha =.5)
axes.plot(x,x+2, color = 'blue')
axes.plot(x,x+3, color= 'green')
```

```python
fig, axes = plt.subplots(dpi= 150)
axes.plot(x,x**2, color = 'red', alpha=.5)
axes.plot(x,x+2, color = 'blue', alpha=.5)
axes.plot(x,x+3, color= 'green', alpha=.5)
```
